## 1.7 Papers group by Size of pages

```python
list_of_columns_using_pages = ['Paper Id', 'Citation', 'Number of Pages', 'Title']

mask1 = df_sms_extraction['Number of Pages'] <= 5
df_papers_until_5_pages = df_sms_extraction[mask1][list_of_columns_using_pages]
df_papers_until_5_pages = df_papers_until_5_pages.sort_values(by=['Number of Pages'])

mask2 = (df_sms_extraction['Number of Pages'] > 5) & (df_sms_extraction['Number of Pages'] <= 15)
df_papers_from_6_to_15_pages = df_sms_extraction[mask2][list_of_columns_using_pages]
df_papers_from_6_to_15_pages = df_papers_from_6_to_15_pages.sort_values(by=['Number of Pages'])

mask3 = (df_sms_extraction['Number of Pages'] > 15) & (df_sms_extraction['Number of Pages'] <= 20)
df_papers_from_16_to_20_pages = df_sms_extraction[mask3][list_of_columns_using_pages]
df_papers_from_16_to_20_pages = df_papers_from_16_to_20_pages.sort_values(by=['Number of Pages'])

mask4 = df_sms_extraction['Number of Pages'] > 20
df_papers_over_20_pages = df_sms_extraction[mask4][list_of_columns_using_pages]
df_papers_over_20_pages = df_papers_over_20_pages.sort_values(by=['Number of Pages'])
```

```python
def macro_wordcloud_content_analysis(path_file, name):
    util.generateWordcloudFromTextFile(path_file, name)
    util.contentAnalysisFromTextFile(path_file)
```

## 2.1 We are creating a text file appended for each analyzed column. 

```python
path_abstracts_file = '../dataset/abstracts.txt'
path_goals_file = '../dataset/goals.txt'
path_conclusions_file = '../dataset/conclusions.txt'
path_futures_file = '../dataset/futures.txt'

path_strong_aspects = '../dataset/strongest.txt'
path_weak_aspects = '../dataset/weaks.txt'
path_scientific_rigor = '../dataset/rigors.txt'
path_paper_relevance = '../dataset/relevances.txt'
path_limitations = '../dataset/limitations.txt'

column_names_df_sms_extraction = df_sms_extraction.columns
abstract_column = column_names_df_sms_extraction[6]
goal_column = column_names_df_sms_extraction[13]
conclusions_column = column_names_df_sms_extraction[20]
strong_aspects_column = column_names_df_sms_extraction[23]
weak_aspects_column = column_names_df_sms_extraction[24]
scientific_rigor_column = column_names_df_sms_extraction[25]
paper_relevance_column = column_names_df_sms_extraction[26]
limitations_column = column_names_df_sms_extraction[27] 
future_column = column_names_df_sms_extraction[28]

util.appendContentsInTextFile(path_abstracts_file, df_sms_extraction[abstract_column])
util.appendContentsInTextFile(path_goals_file, df_sms_extraction[goal_column])
util.appendContentsInTextFile(path_conclusions_file, df_sms_extraction[conclusions_column])
util.appendContentsInTextFile(path_strong_aspects, df_sms_extraction[strong_aspects_column])
util.appendContentsInTextFile(path_weak_aspects, df_sms_extraction[weak_aspects_column])
util.appendContentsInTextFile(path_scientific_rigor, df_sms_extraction[scientific_rigor_column]) 
util.appendContentsInTextFile(path_paper_relevance, df_sms_extraction[paper_relevance_column]) 
util.appendContentsInTextFile(path_limitations, df_sms_extraction[limitations_column])
util.appendContentsInTextFile(path_futures_file, df_sms_extraction[future_column])
```
