```python
path_abstracts_file = '../dataset/abstracts.txt'
path_goals_file = '../dataset/goals.txt'
path_conclusions_file = '../dataset/conclusions.txt'
path_futures_file = '../dataset/futures.txt'

column_names_df_sms_extraction = df_sms_extraction.columns
abstract_column = column_names_df_sms_extraction[6]
goal_column = column_names_df_sms_extraction[12]
conclusions_column = column_names_df_sms_extraction[19]
future_column = column_names_df_sms_extraction[27]

util.appendContentsInTextFile(path_abstracts_file, df_sms_extraction[abstract_column])
util.appendContentsInTextFile(path_goals_file, df_sms_extraction[goal_column])
util.appendContentsInTextFile(path_conclusions_file, df_sms_extraction[conclusions_column])
util.appendContentsInTextFile(path_futures_file, df_sms_extraction[future_column])
```
