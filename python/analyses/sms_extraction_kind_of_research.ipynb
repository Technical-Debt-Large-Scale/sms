```python
import pandas as pd
import util.utilities as util
from collections import Counter
import numpy as np

df_sms_extraction = pd.read_excel('../../dataset/Extraction_form_basic.xlsx')
df_sms_extraction = df_sms_extraction.replace(np.nan, '', regex=True)
```

```python
dict_kind_of_research = {}
list_all_kind_of_research = []
for item in zip( list(df_sms_extraction['Paper Id']), list(df_sms_extraction['Method of research (Case study, Survey, Interview, Experiment, Observation)']), list(df_sms_extraction['Citation']), list(df_sms_extraction['Title']) ):
    id = item[0]
    list_aux = []
    list_of_contents = []
    if (item[1] != ''):
        kind_of_research = str(item[1])
        kind_of_research = ' '.join(kind_of_research.split())
        kind_of_research = kind_of_research.lower()
        list_aux = kind_of_research.split(',')
        citation = str(item[2])
        title = str(item[3])
        title = ' '.join(title.split())
        list_aux2 = []
        for each in list_aux: 
            if each[0] == ' ':
                each = each[1:]
            list_aux2.append(each)
            list_all_kind_of_research.append(each)
        list_of_contents.append(citation)
        list_of_contents.append(title)    
        list_of_contents.append(list_aux2)
        dict_kind_of_research[id] = list_of_contents
print(f'Size of dictionaries of kind_of_research: {len(dict_kind_of_research)}')
dict_kind_of_research
```

```python
Counter(list_all_kind_of_research).most_common()
```
